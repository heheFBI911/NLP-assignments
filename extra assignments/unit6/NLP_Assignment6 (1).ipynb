```python
# Define sample data
data = {
    "text": [
        "I love this product! It's amazing!",  # Positive
        "This is the best experience I've had.",  # Positive
        "The service was okay, nothing special.",  # Neutral
        "I feel indifferent about this.",  # Neutral
        "I hate waiting in long queues.",  # Negative
        "This is the worst experience ever!",  # Negative
        "The food was delicious and well-prepared.",  # Positive
        "The weather today is just normal.",  # Neutral
        "The movie was terrible and boring.",  # Negative
        "The customer support was helpful.",  # Positive
        "The internet speed is average.",  # Neutral
        "My flight got delayed again! So frustrating.",  # Negative
        "I am so happy with my new laptop.",  # Positive
        "It's just another regular day.",  # Neutral
        "This app crashes all the time! So annoying.",  # Negative
    ],
    "sentiment": [
        "Positive", "Positive", "Neutral", "Neutral", "Negative",
        "Negative", "Positive", "Neutral", "Negative", "Positive",
        "Neutral", "Negative", "Positive", "Neutral", "Negative"
    ]
}
```

```python
# Save as CSV file
df.to_csv("sample_sentiment_dataset.csv", index=False)
```

```python
import pandas as pd
import torch
import numpy as np
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
from sklearn.model_selection import train_test_split
from torch.nn.functional import softmax
```

```python
# Encode labels (0=Negative, 1=Neutral, 2=Positive)
label_map = {"Negative": 0, "Neutral": 1, "Positive": 2}
df["label"] = df["sentiment"].map(label_map)
```

```python
# ✅ Step 2: Split the Data (80% Train, 20% Test)
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df["text"].tolist(), df["label"].tolist(), test_size=0.2, random_state=42
)
```

```python
# ✅ Step 3: Convert Data to Hugging Face Dataset Format
train_data = Dataset.from_dict({"text": train_texts, "label": train_labels})
test_data = Dataset.from_dict({"text": test_texts, "label": test_labels})
```

```python
# ✅ Step 4: Load RoBERTa Tokenizer and Model
model_name = "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)
```

```python
# ✅ Step 5: Tokenization Function
def tokenize_function(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True)
```

```python
train_data = train_data.map(tokenize_function, batched=True)
test_data = test_data.map(tokenize_function, batched=True)
```

```python
# ✅ Step 6: Define Training Arguments
training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=5,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    logging_dir="./logs",
    logging_steps=10,
)
```

```python
# ✅ Step 7: Fine-Tune RoBERTa Model
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=test_data,
)

trainer.train()
```

```python
# ✅ Step 8: Evaluate the Model
results = trainer.evaluate()
print("Evaluation Results:", results)
```

```python
# Detect GPU if available, otherwise use CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Move model to the correct device
model.to(device)

# Function for Sentiment Prediction (with proper device handling)
def predict_sentiment(text):
    tokens = tokenizer(text, return_tensors="pt", padding=True, truncation=True).to(device)  # Move to same device
    with torch.no_grad():
        output = model(**tokens)
    probs = softmax(output.logits, dim=1)

    label_map_reverse = {0: "Negative", 1: "Neutral", 2: "Positive"}
    sentiment = label_map_reverse[torch.argmax(probs).item()]

    return sentiment, probs.cpu().numpy()  # Move output to CPU for printing
```

```python
# ✅ Step 10: Test the Model
test_text = "It was bad experience"
sentiment, probabilities = predict_sentiment(test_text)
print(f"Sentiment: {sentiment}")
print(f"Probabilities: {probabilities}")
```
